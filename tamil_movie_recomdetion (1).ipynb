# to find outliers

```python
from scipy import stats

# Select numerical columns
numerical_columns = ['Rating', 'PeopleVote', 'Year', 'Hero_Rating', 'movie_rating', 'content_rating']

# Convert non-numeric values to NaN
for column in numerical_columns:
    concatenated_dataset[column] = pd.to_numeric(concatenated_dataset[column], errors='coerce')

# Calculate z-scores for each numerical column
z_scores = {}
for column in numerical_columns:
    z_scores[column] = stats.zscore(concatenated_dataset[column].dropna())

# Check for extreme z-score values
extreme_z_scores = {}
for column, scores in z_scores.items():
    extreme_z_scores[column] = sum((scores > 3) | (scores < -3))

print("Number of extreme z-scores for each column:")
print(extreme_z_scores)

```

# feature engineerin

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer


df['Genre'] = df['Genre'].fillna('')
df['Actor'] = df['Actor'].fillna('')
df['Director'] = df['Director'].fillna('')


combined_features = (df['Genre'].str.lower() + ' ' + 
                     df['Actor'].str.lower() + ' ' + 
                     df['Director'].str.lower())


tfidf = TfidfVectorizer(stop_words='english')


tfidf_matrix = tfidf.fit_transform(combined_features)

print(tfidf_matrix.shape)
         
```

# Similarity Calculation

```python
from sklearn.metrics.pairwise import linear_kernel


cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

# Recommendation

```python
def recommend_movies(movie_title, data, cosine_sim):
    movie_subset = data[data['MovieName'] == movie_title]
    if len(movie_subset) == 0:
        return "Movie not found in the dataset."
    
    idx = movie_subset.index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  
    movie_indices = [i[0] for i in sim_scores]
    return data['MovieName'].iloc[movie_indices]

recommended_movies = recommend_movies('Ramayya Vasthavayya', df, cosine_sim)


```
